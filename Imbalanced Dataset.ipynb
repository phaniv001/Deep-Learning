## Under Sampling

```python
df_class_no_count, df_class_yes_count = data.Churn.value_counts()

df_class_yes_count, df_class_no_count
```

```python
df_class_yes = data[data['Churn'] == 'Yes']
df_class_no = data[data['Churn'] == 'No']
```

```python
df_class_yes.shape, df_class_no.shape
```

```python
df_class_no_under = df_class_no.sample(df_class_yes_count)

df_class_no_under.shape
```

```python
df_test_under = pd.concat([df_class_no_under, df_class_yes], axis = 0)

df_test_under.shape
```

```python
df_test_under.head()
```

```python
df_test_under.Churn.value_counts()
```

## Oversampling

```python
df_class_yes_over = df_class_yes.sample(df_class_no_count, replace = True)

df_test_over = pd.concat([df_class_yes_over, df_class_no], axis = 0)

df_test_over.Churn.value_counts()
```
